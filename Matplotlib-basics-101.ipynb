```python
x = np.linspace(0,10,100) #.linespace() method gives evenly spaced set of number in the given interval
x
```

#### 1. Line Plot

```python
# Lets try to plot the above data

# Line Plot
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, x**2)
plt.show()
```

#### 2. Scatter Plot

```python
# Create a scatter plot using the same data
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(x,x**2);
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(x,np.sin(x));
```

#### 3. Bar Plot

```python
#lets create a bar plot from a dictionary
fruits = {'Apple':30,'Banana':20,'Straberries':50,'Papaya':15}

fig,ax = plt.subplots(figsize=(8,8))
ax.bar(fruits.keys(), fruits.values());
ax.set(
    title='Fruits in the store',
    xlabel='Type of fruit',
    ylabel='Quantity'
);
```

```python
# Horizontal bar graph
fig,ax = plt.subplots(figsize=(8,8))
ax.barh(list(fruits.keys()), list(fruits.values()));
ax.set(
    title='Fruits in the store',
    xlabel='Quantity',
    ylabel='Type of fruit'
);
```

#### 4. Histograms

```python
# create data and plot a histogram

hist_data = np.random.randn(1000) # Sample of standard normal distribution values
fig, ax = plt.subplots(figsize=(8,8))
ax.hist(hist_data);
```

#### 5. Subplots

multiple plots on the same figure

```python
# Method 1 (Recommended)

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax1.plot(x,x**2);
ax2.scatter(x, np.sin(x));
ax3.bar(fruits.keys(),fruits.values());
ax4.hist(hist_data);
```

```python
# Method 2

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10)) #ax is assigned a tuple with all the axeses
ax[0,0].plot(x,x**2);
ax[0,1].scatter(x, np.sin(x));
ax[1,0].bar(fruits.keys(),fruits.values());
ax[1,1].hist(hist_data);
```

### Pyplot method VS Object Oriented method (Which is better?)
* When you need a simple quick plot is okay to use __pyplot method__.
* And if you have more complex things to do and have more control over your plots, DEFINIETLY use __OO method__.

```python
#lets get some plots about above 50 people

heart_diseas_abv50 = heart_diseas[heart_diseas['age']>50]
len(heart_diseas_abv50)
```

```python
heart_diseas.head()
```

```python
heart_diseas_abv50.head()
```

```python
# using pyplot method
heart_diseas_abv50.plot(x='age',y='chol',c='target',kind='scatter',figsize=(10,8));
```

```python
# Now lets use OO method
fig, ax = plt.subplots(figsize=(10,8))

# plot the data
scatter_plt = ax.scatter(x=heart_diseas_abv50['age'],
                         y=heart_diseas_abv50['chol'],
                         c=heart_diseas_abv50['target']
                        );

# customize
ax.set(title='Relationship Cholestarol Levels and Age',
       xlabel='Age',
       ylabel='cholestarol Level',
      xlim=[45,100]);

# add a legend
ax.legend(*scatter_plt.legend_elements(),title='target')

# Add a mean line
ax.axhline(y=heart_diseas_abv50['chol'].mean(), linestyle='--');
```

```python
heart_diseas_abv50.head()
```

```python
# Create two subplots share the same x axis
fig, (ax0,ax1)= plt.subplots(nrows=2, ncols=1, figsize=(15,20),sharex=True);

# Plot the ax0
scatter= ax0.scatter(x=heart_diseas_abv50['age'],y=heart_diseas_abv50['chol'],c=heart_diseas_abv50['target']);

# customize
ax0.set(title='Cholesterol vs Age',ylabel='Cholestarol Level');

# adding the legend
ax0.legend(*scatter.legend_elements(), title='Target');

#adding amean line
ax0.axhline(heart_diseas_abv50['chol'].mean(),linestyle='--');


# age vs thalach

scatter1 = ax1.scatter(x=heart_diseas_abv50['age'],y=heart_diseas_abv50['thalach'],c=heart_diseas_abv50['target'])

ax1.set(title='Age VS Thalach', xlabel='Age', ylabel='Thalach Level')

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(y=heart_diseas_abv50['thalach'].mean(), linestyle='--');

# Add a figure title

fig.suptitle('Heart Diseas Analysis',fontsize=32, fontweight='bold');
```

```python
# Setting the style
plt.style.use('seaborn-whitegrid')

# Now lets use OO method
fig, ax = plt.subplots(figsize=(10,8))

# plot the data
scatter_plt = ax.scatter(x=heart_diseas_abv50['age'],
                         y=heart_diseas_abv50['chol'],
                         c=heart_diseas_abv50['target'],
                         cmap='winter' #its going to change the color theme
                        );

# visit matplot lib color tutorial

# customize
ax.set(title='Relationship Cholestarol Levels and Age',
       xlabel='Age',
       ylabel='cholestarol Level',
      xlim=[45,100]);

# add a legend
ax.legend(*scatter_plt.legend_elements(),title='target')

# Add a mean line
ax.axhline(y=heart_diseas_abv50['chol'].mean(), linestyle='--');
```

```python
# Setting the style
plt.style.use('seaborn-whitegrid')

# Create two subplots share the same x axis
fig, (ax0,ax1)= plt.subplots(nrows=2, ncols=1, figsize=(15,20),sharex=True,);

# Plot the ax0
scatter= ax0.scatter(x=heart_diseas_abv50['age'],
                     y=heart_diseas_abv50['chol'],
                     c=heart_diseas_abv50['target'],
                    cmap='winter');

# customize
ax0.set(title='Cholesterol vs Age',ylabel='Cholestarol Level');

ax0.set_xlim([50,80]) # change the x axis limits


# adding the legend
ax0.legend(*scatter.legend_elements(), title='Target');

#adding amean line
ax0.axhline(heart_diseas_abv50['chol'].mean(),linestyle='--');


# age vs thalach

scatter1 = ax1.scatter(x=heart_diseas_abv50['age'],
                       y=heart_diseas_abv50['thalach'],
                       c=heart_diseas_abv50['target'],
                      cmap='winter')

ax1.set(title='Age VS Thalach', xlabel='Age', ylabel='Thalach Level')

ax1.set_ylim([60,200]) # change the y axis limits

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(y=heart_diseas_abv50['thalach'].mean(), linestyle='--');

# Add a figure title

fig.suptitle('Heart Diseas Analysis',fontsize=32, fontweight='bold');
```
