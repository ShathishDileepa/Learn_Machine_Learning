### Markdown Syntax

Markdown is simply a way of creating web pages without very syntaxy html code, very good for blogging. It wil turn to regular HTML after running trough a Markdown engine.

Here are some important pointer about __Markdown Syntax__.

* You can create Headers just like in HTML, and yes there are 6 type of headers just like in HTML.
* Starting from h1(which is the biggest) to h6(wich is the smallest).
* Depending on which type of header you can type the number of hastags('#') corresponding to the header and leave a space and type the content of the header. As shown Below


__# Header 1__

__## Header 2__ 

__## Header 3__

__### Header 4__

__#### Header 5__

__##### Header 6__

* You can add a Horizental Ruler by typing three '_'s.

ex : '___'
___

* And for __Bold a text__ you can either start the piece of text with two '\*' or two '\_'s and finish with two '\*' or two '\_'s.

ex : \*\*Bold Text** or \_\_Bold Text__ >> __Bold Text__ 

* To __Italic a text__ you can either start the piece of text with '\*' or '\_' and finish with '\*' or '\_'.

ex : \*Italic Text* or \_Itallic Text_ >> _Itallic Text_ 

* To create an __Unordered list__ you can start each list item with a '\*'.

* To create a __Ordered List__ you can simply number the list items.

* And you can nest them as same as in HTML.

ex: 

* UOL item 1
* UOL item 2
    1. Nested OL item 1
    2. Nested OL item 2
* UOL item 3

### Manipulate Data with Pandas

```python
# We can add colums to our data frames using python lists, Pandas series, or constant value.

new_car_sales
```

```python
# Lets add a constant value to the dataframe
new_car_sales['Wheels'] =4
new_car_sales
```

```python
# Lets try to add a previous owner column by using a python list
previous_owners = [1,1,2,1,3,2,4,5] #only has 8 elements
```

```python
new_car_sales['Previous owners']=previous_owners

# Gives an error... Mmmm why is that
# if we are using a python list it has to have the same number of elements as the number of rows dataframe has
```

```python
previous_owners = [1,1,2,1,3,2,4,5,1,2]
new_car_sales['Previous owners']=previous_owners
new_car_sales

# now it works
```

```python
# We can do column operation as follows as well
new_car_sales = pd.read_csv('./data/car-sales.csv')
new_car_sales
```

```python
new_car_sales['Fuels per (KM)'] = pd.Series([5,4.5,3.5,2,4,4,6,7.5,2,3])
new_car_sales
```

```python
# We can perform following kind of operations and populate the dataframe with derived values
new_car_sales['Total Fuel Spent']=new_car_sales['Odometer (KM)']*new_car_sales['Fuels per (KM)']
new_car_sales
```

```python
new_car_sales['Total Fuel Spent (L)']=new_car_sales['Odometer (KM)']*new_car_sales['Fuels per (KM)']
new_car_sales
```

```python
# we can drop columns as follows
# axis =1 means column
new_car_sales.drop('Total Fuel Spent', inplace=True, axis=1)
new_car_sales
```

```python
# And also we can apply functions to each data in columns as well

new_car_sales['Odometer (Miles)'] = new_car_sales['Odometer (KM)'].apply(lambda x: x*0.6)
```

```python
new_car_sales
```

#### Shuffling the data around 

```python
# We can shuffle and get randomize sample using .sample() method
# Frac here means fraction and it means which percentage of data need to be shuffled and sampled
new_car_sales.sample(frac=0.25, axis=0)
```

```python
new_car_sales_shuffled = new_car_sales.sample(frac=1, axis=0)
new_car_sales_shuffled
```

```python
# we can reset this shuffled state as well

new_car_sales_shuffled.reset_index()
```

```python
# to get rid of that index column
new_car_sales_shuffled.reset_index(drop=True)
```
